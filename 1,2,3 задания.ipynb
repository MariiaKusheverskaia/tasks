### Перебор параметров по сетке

```python
for i in tqdm(range(50)):
    for j in range(50):
        for k in range(50):
            for p in range(50):
                b = 1 + (4 - 1) / 49 * i
                eps = 0 + (0.5 - 0) / 49 * j
                mu = 0 + (0.05 - 0) / 49 * k
                nu0 = 0 + (L_stat[0]/(1.3 * Y_stat[0]) - 0) / 49 * p
                
                M = [0]*len(Y_stat)
                x = [0]*len(Y_stat)
                sigma = [0]*len(Y_stat)
                alpha = [0]*len(Y_stat)
                nu = [0]*len(Y_stat)
                alpha = [0]*len(Y_stat)
                f = [0]*len(Y_stat)
                Y = [0]*len(Y_stat)
                
                M[0] = 1.3 * Y_stat[0]
                nu[0] = nu0
                sigma[0] = J_stat[0] / (b * M[0])
                alpha[0] = 1 - eps - mu / sigma[0]
                x[0] = L_stat[0] / M[0]
                f[0] = 1 - (1 - alpha[0] * x[0]/nu[0]) ** (1 / alpha[0])
                Y[0] = M[0] * f[0]
                is_ok = True
                for t in range (1, 19):
                    M[t] = M[t-1] * (1 - mu) + J_stat[t-1] / b
                    sigma[t] = J_stat[t] / (b * M[t])
                    nu[t] = nu[t-1] * (1 - eps * sigma[t-1])
                    alpha[t] = 1 - eps - mu / sigma[t]
                    x[t] = L_stat[t] / M[t]
                    f[t] = 1 - (1 - alpha[t] * x[t]/nu[t]) ** (1 / alpha[t])
                    Y[t] = M[t] * f[t]
                    if (sigma[t] > 0.2) or (nu[t] >= x[t]):
                        is_ok = False
                        break
                if is_ok:
                    #print(b, eps, mu, nu0)
                    b_list.append(b)
                    eps_list.append(eps)
                    mu_list.append(mu)
                    nu_list.append(nu0)
```
